```python
##Now I can use the whole dataframe to train on instead of the sample

def X_y_train(training, rhs, depvar):
    X_train = training[rhs].copy()
    
    y_train = training[depvar]

    X_columns = X_train.columns

    # imputation
    imputer = Imputer(strategy='mean')
    X_train = imputer.fit_transform(X_train)


    return X_train, y_train, X_columns, imputer
```

```python
def fit_gbc(X_train, y_train):
    print "Training model"
    t0 = time.time()

    gbc_final = GradientBoostingClassifier(criterion='friedman_mse', init=None,
              learning_rate=0.05, loss='deviance', max_depth=5,
              max_features=None, max_leaf_nodes=None,
              min_impurity_split=1e-07, min_samples_leaf=1,
              min_samples_split=2, min_weight_fraction_leaf=0.0,
              n_estimators=200, presort='auto', random_state=None,
              subsample=1.0, verbose=0, warm_start=False)

    gbc_final.fit(X=X_train, y=y_train.as_matrix()[:,0])
    t1 = time.time()
    print "Total runtime {} minutes".format((t1-t0)/60)
    return gbc_final
```

```python
depvar = ['signupdv_60']
rhs = list(set(training.columns) - set(depvar))

X_train, y_train, X_columns, imputer = X_y_train(training, rhs, depvar)

gbc_final = fit_gbc(X_train, y_train)
```
